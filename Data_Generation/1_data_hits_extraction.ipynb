# Data Hits Extraction

Para la extracción de los datos se ha desarrollado un objeto (Music_Data), con varias funciones.

El objeto ofrece un total de 5 funciones.

1. Get_Billboard_Data -> Dado un rango de fechas se extraen todos los hit Hot-100 de billboard en ese rango y se añaden diferentes caracteristicas de cada una de las canciones obtenidas a través de la API de spotify (spotipy)

2. Get_Billboard_Data_Lyrics -> Esta función es igual a la anterior, y añade la letra de cada una de las canciones. Se crearon 2 funciones, ya que en un principio se usaba esta función. Pero el proceso es bastante lento en comparación con el anterior, debido al funcionamiento de la API de extracción de canciones (Lyric Genius). Como existen muchas duplicidades en la extracción se pensó que seria mejor hacer una extracción sin letras de canciones y una vez eliminados todos los duplicados añadir las letras, ahorrando mucho tiempo en el proceso. Se ha decido mantener la función.

3. Get_User_Playlist_Data -> Función que permite obtener, para un usuario de Spotify dado, todas las canciones de todas sus playlist publicas, con las características de las mismas.

4. Get_User_Playlist_Data_Lyrics -> Función igual a la anterior, que añade a cada una de las canciones, la letra de la misma. Por lo mismo que comentado en el punto 2. Se han realizado 2 funciones.

5. Get_Lyrics -> Función que dado un dataframe, de los tipos generados con Get_Billboard_Data o Get_User_Playlist_Data, añade la letra de las canciones, esta función se aplicará cuando se hallan eliminado las duplicidades, y por tanto será mucho más rápido el proceso.


La extracción se ha ido realizando por año desde 1958 (Desde que se existen registros de Hot-100) hasta Agosto de 2020. Se ha realizado una extraccion por año para evitar errores en la extraccion debido a conexion ya que si el rango de fechas es demasiado grande, el tiempo de duración de extraccion sera mas alto y puede producir problemas.

Como se han realizado extracciones desde 1958 hasta 2020, se han realizado un total de 62 extracciones.
Algunas se realizaron con Get_Billboard_Data_Lyrics y otras con Get_Billboard_Data, para agilizar el proceso.
